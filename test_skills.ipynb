```python
import rioxarray as rxr
import matplotlib.pyplot as plt
import earthpy.plot as ep

# Open the pre_DTM_hill.tif hillshade layer
hillshade_path = "colorado-flood/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM_hill.tif"
hillshade = rxr.open_rasterio(hillshade_path)

# Open the DTM data
dtm_pre_path = "colorado-flood/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM.tif"  # Update with the correct path to your DTM data
dtm_pre = rxr.open_rasterio(dtm_pre_path)

# Plot both layers on the same figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the hillshade
ep.plot_bands(hillshade,
 cmap="gray",  # Set the colormap to Greys
              ax=ax)

# Plot the DTM with adjusted opacity
ep.plot_bands(dtm_pre,
              cmap='viridis',  # Set the colormap to viridis
              alpha=0.5,  # Adjust the opacity of the DTM
              ax=ax)

# Add titles and legend
ax.set_title("Overlay of DTM on DTM Hillshade")
ax.legend(["Hillshade", "DTM"])

# Show the plot
plt.show()
```
